```python
# Cell 1 — Imports
from src.chunking import chunk_text
```

```python
# Cell 2 — Load extracted text
with open("data/fabric-data-engineering.txt", "r", encoding="utf-8") as f:
    full_text = f.read()

print(f"Full text length: {len(full_text):,} characters")
```

```python
# Cell 3 — Slice a small sample to test the pipeline cheaply
# Validate chunk -> embed -> index -> query on a slice before paying
# embedding/indexing cost on the full document.
sample_text = full_text[0:200000]
print(f"Sample length: {len(sample_text):,} characters")
```

```python
# Cell 4 — Chunk the sample
chunks = chunk_text(sample_text, source="fabric-data-engineering", chunk_size=1500, chunk_overlap=150)
print(f"Produced {len(chunks)} chunks")
```

```python
# Cell 5 — Inspect the first few chunks
for c in chunks[:10]:
    print(f"ID:      {c['id']}")
    print(f"Source:  {c['source']}")
    print(f"Index:   {c['chunk_index']}")
    print(f"Length:  {len(c['content'])} chars")
    print(f"Content:\n{c['content']}")
    print("-" * 60)
```
